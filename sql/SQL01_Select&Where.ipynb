### 문제 3>  city 테이블에서 population이 800만 ~ 1000만 사이인 도시 데이터를 인구수순으로 내림차순하세요.

### 문제4> country 테이블에서 1940 ~ 1950년도 사이에 독립한 국가들을 조회하고 독립한 년도 순으로 오름차순하세요.

```python
# country 테이블 구성 확인
Query = """
    SELECT *
    From country
    
"""
```

```python
df = pd.read_sql(Query, db)
df
```

```python
Query = """
    SELECT Code, Concat(Name, "(", IndepYear, ")") as NameIndep, Population
    From country
    Where IndepYear between 1940 and 1950
    Order by IndepYear
"""
```

```python
df = pd.read_sql(Query, db)
df
```

### 문제 5> contrylanguage 테이블에서 스페인어, 한국어, 영어를 95% 이상 사용하는 국가 코드를 Percentage로 내림차순하여 아래와 같이 조회하세요.

```python
Query = """
    Select *
    From countrylanguage
"""
```

```python
df = pd.read_sql(Query, db)
df
```

```python
Query = """
    SELECT CountryCoDE, Language, Percentage
    From countrylanguage
    Where Language in ("English", "Spanish", "Korean")
    and Percentage >= 95
    Order by Percentage DESC
"""
```

```python
df = pd.read_sql(Query, db)
df
```
